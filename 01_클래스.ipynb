```python
# class는 상태 관리에 유용하지만, 꼭 사용해야하는 방법은 아님. 함수로도 충분하므로 개인의 선택.

# 아래는 함수임.
result = 0

def add(num):
    global result
    result += num # 전역변수 사용해서 계속 누적 합이 됨.
    return result

print(add(3))
print(add(4))
print(add(5))
```

```python
# 클래스 만들기

class Calculator:
    # 클래스의 상태값 = 프로퍼티
    def __init__(self): # self는 초깃값
        self.result = 0
    
    # 클래스의 함수 = 메서드
    def add(self, num):
        self.result += num
        return self.result

cal1 = Calculator()
print(cal1.add(3))
print(cal1.add(4))

cal2 = Calculator()
print(cal2.add(5))
print(cal2.add(6))
```

```python
class Calculator:
    # 클래스의 상태값 = 프로퍼티
    def __init__(self): # self는 초깃값
        self.result = 0
    
    # 클래스의 함수 = 메서드
    def add(self, num):
        self.result += num
        return self.result
    
    def sub(self, num):
        self.result -= num
        return self.result
    
    def mul(self, num):
        self.result *= num
        return self.result
    
    def div(self, num):
        self.result /= num
        return self.result
    
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(10))
print(cal1.sub(2))
print(cal1.mul(5))
print(cal1.div(4))

print()

print(cal2.add(5))
print(cal2.mul(10))
print(cal2.sub(10))
print(cal2.div(5))
```

```python
# 객체에 연산할 숫자 지정하기

class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    
    def add(self):
        result = self.first + self.second
        return result
    
cal3 = FourCal()

print(type(cal3))
print(cal3.setdata(4,2))
print()

print(cal3.first)   # 프로퍼티
print(cal3.second)  # 프로퍼티
print()

print(cal3.add())   # 메소드
```

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    
    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    
cal4 = FourCal()
cal4.setdata(8, 4)

print(cal4.add())
print(cal4.sub())
print(cal4.mul())
print(cal4.div())
```
