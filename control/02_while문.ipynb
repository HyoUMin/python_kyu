```python
# while문 기본 구조
# while 조건문:
#     수행할 문장1
#     수행할 문장2
#     수행할 문장3  
#     ...
```

```python
treehit = 0

while treehit < 10:
    treehit += 1
    print(f"나무를 {treehit}번 찍었습니다.")
    if treehit == 10:
        print("나무 넘어갑니다")
```

```python
number = 0
prompt = """
1. ADD
2. DEL
3. LIST
4. QUIT
"""

while number != 4:
    print(prompt)
    number = int(input("1에서 4까지의 숫자 중 하나를 입력해주세요"))
```

```python
# while문 탈출하기

coffee = 5
money = 300

while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffee -= 1
    print(f"남은 커피는 {coffee}개입니다")
    if coffee == 0:
        print("커피 재고 소진입니다. 판매를 중지합니다.")
        break
```

```python
coffee = 10

while True:
    money = int(input("돈을 지불해주세요."))
    if money == 300:
        print("커피를 드리겠습니다.")
        coffee -= 1
    elif money > 300:
        print(f"거스름돈 {money - 300}원을 드리겠습니다.")
        coffee -= 1
    else:
        print("돈을 다시 돌려주고, 커피를 드리지않습니다.")
        print(f"현재 남은 커피는 {coffee}개입니다.")
    
    if coffee == 0:
        print("커피 재고가 없습니다. 판매 종료합니다")
        break
```

```python
# while문의 처음으로 돌아가가 - continue
# 1~10 중에서 홀수만 출력하기

num = 0
while num < 10:
    num += 1
    if num % 2 == 0:
        continue
    print(num)
```
