```python
# for문의 기본구조
# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할 문장1
#     수행할 문장2
#     ...

num_list = ["one", "two", "three"]
for num in num_list:
    print(num)
```

```python
a = [(1, 2), (3, 4), (5, 6)]

for i in a:
    print(i)
```

```python
tuple_list = [(1, 2), (3, 4), (5, 6)]

for a in tuple_list:
    print(a)
```

```python
a, b = (1, 2)
a, b = 1, 2
[a, b] = [1, 2]
[a, b] = (1, 2)
[a, *b] = (1, 2, 3)
print(a, b)
# 언패킹 참고할 것
```

```python
marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
    number += 1
    if mark >= 60:
        print(f"{number}번 학생은 합격입니다.")
    else:
        print(f"{number}번 학생은 불합격입니다.")
```

```python
# for문과 continue문

marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
    number += 1
    if mark < 60:
        continue
    else:
        print(f"{number}번 학생은 합격입니다.")
```

```python
# for문과 함께 range함수 사용하기

a = range(10)
print(list(a))

b = range(0, 10)
print(list(b))

c = range(2, 10)
print(list(c))
```

```python
result = 0
for number in range(1, 11):
    result += number
    print(result)
print(f"최종값: {result}")
```

```python
marks = [90, 25, 67, 45, 80]

for number in range(len(marks)):
    if marks[number] < 60:
        continue
    print(f"{number+1}번 학생은 합격입니다.") # number에 +1 안해주면 0,1,2,3,4 로 원하는 값이 출력되지 않음.
```

```python
# 중첩 for문
for i in range(2, 10):
    for j in range(1, 10):
        print(i * j, end=" ")
    print("")
```

```python
# 중첩 for문
for i in range(2, 10):
    print(f"{i}단")
    for j in range(1, 10):
        print(f"{i} X {j} = {i * j}")
    print("")
```

```python
for i in range(1,6):
    for j in range(5 -i):
        print(" ", end="")
    for k in range(i):
        print("*", end="")
    print()
```

```python
for i in range(5, 0, -1):
    print("*" * i)
```

```python
# 리스트 컴프리헨션 사용

a = [1, 2, 3, 4]
result = []
for num in a:
    result.append(num * 3)
print(result)
# -------------------------------------

b = [1, 2, 3, 4]
result1 = [num1 * 3 for num1 in b]
print(result1)
```

```python
# for문과 break문 사용

for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

```python
# enumerate() 함수 사용하기. 인덱스 번호 가져오는 것이 가능해짐. 자주 활용할 예정임.

fruits = ["apple", "banana", "orange"]
for i, fruit in enumerate(fruits):
    print(f"{i}번째: {fruit}")
```

```python
# zip()함수로 여러 리스트 함께 순회하기.

names = ["홍길동", "유재석", "강호동"]
scores = [60, 70, 80]

for name, score in zip(names, scores):
    print(f"{name} 학생은 {score}점입니다.")
```

```python
# 3개이상의 리스트를 zip()으로 다루기

names = ["홍길동", "유재석", "강호동"]
korean_scores = [60, 70, 80]
english_scores = [80, 90, 70]

for name, korean_score, english_score in zip(names, korean_scores, english_scores):
    print(f"{name} 학생은 국어 {korean_score}점, 영어 {english_score}점입니다.")
```
