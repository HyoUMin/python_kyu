```python
# 성적이 90점 이상이면 A학점
# 성적이 90점 미만이면 80점 이상이면 B학점
# 성적이 80점 미만이면 70점 이상이면 C학점
# 성적이 70점 미만이면 60점 이상이면 D학점
# 60점 미만이면 F학점 출력

score = 67

if score >= 90:
    print("A학점")
elif score < 90 and score>= 80:
    print("B학점")
elif score < 80 and score >= 70:
    print("C학점")
elif score < 70 and score >= 60:
    print("D학점")
else:
    print("F학점")
```

```python
# match - case 구문
# match 변수
#   case 패턴1:
#       수행할 문장
#   case 패턴2:
#       수행할 문장
```

```python
grade = input("A ~ F 중 하나의 값을 입력하시오")

match grade:
    case "A" | "B" | "C":
        print("합격입니다.")
    case _: # 그 외라는 표현
        print("불합격입니다.")
```
