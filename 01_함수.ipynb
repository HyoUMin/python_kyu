```python
# 파이썬 함수의 구조

# def 함수이름(매개변수1, 매개변수2, ...):    매개변수 = 파라미터
#     수행할 문장1
#     수행할 문장2
#     ...

def add(a, b):
    return a + b
```

```python
result = add(10, 20)
print(result)
```

```python
# 1. 일반적인 함수: 입력값이 있고, 반환값이 있다.

def add(a, b):
    result = a + b
    return result

sum = add(3, 4)
print(sum)
```

```python
# 2. 입력값은 없고, 반환값은 있는 함수

def say():
    return "Hi"

a = say()
print(a) 
```

```python
# 3. 입력값은 있고, 반환값은 없는 함수
def add(a, b):
    print(f"{a} + {b} = {a+b}")
    
sum = add(3, 4)
print(sum)
```

```python
# 4. 입력값과 반환값 모두 없는 함수

def say():
    print("Hi")
    
a = say()
print(a)
```

```python
# 매개변수(parameter)를 지정하여 호출
def sub(a, b):
    return a - b

result = sub(a = 7, b = 3)
print(result)
```
