```python
#  파일 생성하기

f = open("새 파일.txt", "w")
f.close()
```

```python
f = open("새 파일1.txt", "w", encoding = "utf-8") # f는 변수 아님. 객체로 생각해야 함. 파일 객체라고도 불림
for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
# 파일 읽기
f = open("./새 파일1.txt", "r", encoding = "utf-8")
line = f.readline() # 1줄 읽기

print(line)
f.close()
```

```python
# 파일 읽기
f = open("./새 파일1.txt", "r", encoding = "utf-8")

# 전체 라인 읽기
while True:
    line = f.readline()
    if not line:
        break
    print(line)
f.close()
```

```python
# readlines 사용 - 전체 읽기(리스트 형태로 출력)
f = open("./새 파일1.txt", "r", encoding = "utf-8")

lines = f.readlines()
print(lines)
f.close()
```

```python
# 파일에 새로운 내용 추가하기

f = open("./새 파일1.txt", "a", encoding = "utf-8")
for i in range(21, 31):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
# with문과 함께 사용하기.
with open("foo.txt","w") as f:
    f.write("Life is too short, you need python")
```
