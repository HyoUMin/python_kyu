```python
# 파이썬 함수의 구조

# def 함수이름(매개변수1, 매개변수2, ...):    매개변수 = 파라미터
#     수행할 문장1
#     수행할 문장2
#     ...

def add(a, b):
    return a + b
```

```python
result = add(10, 20)
print(result)
```

```python
# 1. 일반적인 함수: 입력값이 있고, 반환값이 있다.

def add(a, b):
    result = a + b
    return result

sum = add(3, 4)
print(sum)
```

```python
# 2. 입력값은 없고, 반환값은 있는 함수

def say():
    return "Hi"

a = say()
print(a) 
```

```python
# 3. 입력값은 있고, 반환값은 없는 함수
def add(a, b):
    print(f"{a} + {b} = {a+b}")
    
sum = add(3, 4)
print(sum)
```

```python
# 4. 입력값과 반환값 모두 없는 함수

def say():
    print("Hi")
    
a = say()
print(a)
```

```python
# 매개변수(parameter)를 지정하여 호출
def sub(a, b):
    return a - b

result = sub(a = 7, b = 3)
print(result)

result1 = sub(b = 7, a = 3)
print(result1)

print("안녕","바이", sep = "---")
```

```python
# 입력값이 몇 개가 될지 모르는 경우 -> 튜플
def add_many(*args):
    result = 0
    print(type(args))
    print(args)
    for i in args:
        result += i
    return result

result = add_many(1, 2, 3, 4, 5)
print(result)
print() 
```

```python
# 입력값이 여러 개일 때, 더하기와 곱하기를 선택할 수 있는 함수
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result += i

    elif choice == "mul":
        result = 1
        for i in args:
            result *= i

    return result
result1 = add_mul("add", 1, 2, 3, 4, 5)
print(result1)

result2 = add_mul("mul", 1, 2, 3, 4, 5)
print(result2)
```

```python
# 키워드 매개변수, kwargs -> 딕셔너리

def print_kwargs(**kwargs):
    print(kwargs)

print_kwargs(a = 1)
print_kwargs(name = "foo", age = 20, job = "student")
```

```python
def create_profile(**kwargs):
    print("===== 프로필 =====")
    for key, value in kwargs.items():
        print(f"{key}: {value}")

create_profile(이름 = "김철수", 취미 = "축구", 나이 = 24)
```

```python
# 함수의 반환 값은 언제나 하나이다.

def add_and_mul(a, b):
    return a + b, a * b # 하나의 튜플 형태로 반환 

result = add_and_mul(3, 4)
print(result)
```

```python
def add_and_mul(a, b):
    return a + b
    return a * b # 이 부분은 실행되지 않음

result = add_and_mul(3, 4)
print(result)
```

```python
def say_nick(nick):
    if nick == "학생":
        return
    print(f"나는 {nick}입니다.") 

say_nick("선생")
say_nick("학생")
```

```python
# 매개변수 초깃값 설정하기

def say_myself(name, age, man = True):
    print(f"너의 이름은 {name}입니다.")
    print(f"너의 나이는 {age}입니다.")
    
    if man:
        print("남자입니다.")
    else:
        print("여자입니다.")


say_myself("김철수", 24)
print()
say_myself("김영희", 22, False)
```

```python
# 함수 안에서 선언한 변수의 효력 범위

a = 1
def vartest(a):
    a += 1
    
vartest(a)
print(a)
```

```python
a = 1
def vartest():
    global a
    a += 1

vartest()
print(a)
```

```python
# lambda 예약어

# 함수_이름 = lambda 매개변수1, 매개변수2, ... : 표현식
def add(a, b):
    return a + b

add = lambda a, b: a + b
```

```python
# 함수의 독스트링(DOCSTRING)

def add(a, b):
    """
    두 수의 합을 반환하는 함수

    Parameters:
    a (int, float): 첫 번째 숫자
    b (int, float): 두 번째 숫자
    
    Returns:
    int, float: 두 수의 합
    """
    return a + b

print(add.__doc__)  
```
