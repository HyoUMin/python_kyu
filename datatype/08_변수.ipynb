```python
a = [1, 2, 3]
print(id(a)) # 주소랑 비슷한 개념의 고유값

b = [1, 2, 3]
print(id(b))

a.append(4)
print(a)
print(id(a))
```

```python
a = [1, 2, 3]
b = a[:] # 복제한다는 개념임.

print(id(a))
print(id(b))

```

```python
# 변수를 만드는 여러가지 방법

a, b = ["python", "like"] # 언패킹
print(a)
print(b)
```

```python
a, b = "python", "like"
print(a)
print(b)
print(type(a))
```

```python
a = b = "python"
print(a)
print(b)
print(type(a))
```
