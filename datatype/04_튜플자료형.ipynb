```python
#  튜플은 어떻게 만들까요? 튜플은 변경 X

a = ()
b = (1,) # 하나의 요소일 때 콤마 필수
c = (1, 2, 3)
d = 1, 2, 3 # 괄호 생략 가능
e = ("a", "b", ("ab", "cd"))
f = 1,
g = 1

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
```

```python
# 튜플 요소 값을 삭제할 때

a = (1, 2, "a", "b")
del  a[0]
```

```python
# 튜플 값을 변경할 때

a = (1, 2, "a", "b")
a[0] = "c"
```

```python
a  = (1, 2)
a = a * 2
print(a)

# a 튜플값이 변경된 것이 아니라 재할당한 과정
```

```python
a = 10
print(id(a))

a = 20
print(id(a))

# 메모리 관점에서 a 값이 변경된 것이 아니라 재할당된 것
```

```python
a = [1, 2, 3]
print(id(a))

a.append(4)
print(a)
print(id(a))
```
