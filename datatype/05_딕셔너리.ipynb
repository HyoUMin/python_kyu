```python
# 딕셔너리 
dic = {"name" : "ken", "phone" : "010-1234-5678", "birth" : "01-01"}
print(dic)
print(type(dic))
```

```python
# 딕셔너리 쌍 추가하기

a = {1: "a"}
a[2] = "b"  # 2는 인덱스가 아니고, 키임. 딕셔너리는 {키-밸류}

print(a)
a["name"] = "ken"
print(a)

a[(1,2)] = "hobby"
print(a)

a[[3,4]] = "job" # 리스트인 [3, 4]는 append를 통해 변경가능하므로 키 값으로 사용못함
print(a)
```

```python
# 딕셔너리 요소 삭제

print(a)

del a[1]

del a["name"]
print(a)
```

```python
# 딕셔너리에서 키 값을 사용하여 밸류 값을 얻어내기

profile = {
    "name" : "홍길동",
    "age" : 25
}

profile["name"]
```

```python
# 딕셔너리 관련 함수
# 1. key 리스트 만들기 - keys

profile = {
    "이름": "홍길동",
    "나이": 25,
    "번호": "010-1234-5678"
}

print(profile.keys())

# 2. value 리스트 만들기 - values

profile = {
    "이름": "홍길동",
    "나이": 25,
    "번호": "010-1234-5678"
}

print(profile.values())

# 3. key:value 쌍 얻기 - items

profile = {
    "이름": "홍길동",
    "나이": 25,
    "번호": "010-1234-5678"
}

print(profile.items())

# 4. key, value 쌍 모두 지우기 - clear

profile = {
    "이름": "홍길동",
    "나이": 25,
    "번호": "010-1234-5678"
}

print(profile.clear())

# 5. key로 value 얻기 - get

profile = {
    "이름": "홍길동",
    "나이": 25,
    "번호": "010-1234-5678"
}

print(profile.get("이름")) # profile["이름"]과 동일한 기능

# 해당 key가 딕셔너리 안에 있는지 조사하기(명령어 O - 함수 아님) - in

profile = {
    "이름": "홍길동",
    "나이": 25,
    "번호": "010-1234-5678"
}

print("이름" in profile)

# key로 value 얻기 - pop

profile = {
    "이름": "홍길동",
    "나이": 25,
    "번호": "010-1234-5678"
}

temp = profile.pop("이름")
print(temp)
print(profile)
```
